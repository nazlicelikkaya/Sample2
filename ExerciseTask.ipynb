# Task 04. What is the total value of sales in EUR?

```python
df_sales1 = df_sales.copy()
df_currency1 = df_currency.copy()
```

```python
df_sales1.head()
```

```python
df_currency1.head()
```

```python
pd.to_datetime(df_sales1.date)
```

```python
df_sales1.date = pd.to_datetime(df_sales1.date) #date i cevirdik
```

```python
pd.to_datetime(df_currency.date)
```

```python
df_currency1.date = pd.to_datetime(df_currency1.date)
```

```python
df_table = pd.merge(df_sales1, df_currency1, how = "inner", on = ["currency","date"])
df_table.head()
```

```python
df_sales1.info()
```

```python
df_currency1.info()
```

```python
df_table["currency"].value_counts()
```

```python
df_table["Euro_total"] = df_table["price"] / df_table["rate"]
df_table.head()
```

```python
df_table.head()
```

```python
df_table["Euro_total"].sum()
```
